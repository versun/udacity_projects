**任务3.2：**分析电影净利润（票房-成本）随着年份变化的情况，并简单进行分析。

```python

movie_data['profit'] = movie_data['revenue_adj'] - movie_data['budget_adj']

plt.figure(figsize=[20,20])

plt.subplot(3,1,1)  #每年总利润趋势
sb.pointplot(data = movie_data, x = 'release_year',y = 'profit',markers='o',estimator=np.sum, scale=1,errwidth=0)
plt.xticks(rotation = 90);
plt.xlabel('Year', fontsize=12)
plt.ylabel('Profit   /1 billion', fontsize=12)

plt.subplot(3,1,2)  #每年总发行量趋势
sb.pointplot(data=movie_data,x=movie_data['release_year'].value_counts().index,y=movie_data['release_year'].value_counts().values)
plt.xticks(rotation = 90);
plt.xlabel('Year', fontsize=12)
plt.ylabel('Movie Counts', fontsize=12)

plt.subplot(3,1,3)  #每年平均利润趋势
sb.pointplot(data=movie_data,x='release_year',y='profit',estimator=np.mean,scale=1,errwidth=0)
plt.xticks(rotation = 90);
plt.xlabel('Year', fontsize=12)
plt.ylabel('Average profit', fontsize=12)
'''
由3个图表可知，
1.利润随着年份的增加呈上升趋势，20世纪开始爆发性增长
2.每年发行量和利润成正比关系
3.每年每部的平均利润，20世纪以来趋于稳定，较上世纪相比，无明显增长
''';
```

**[选做]任务3.3：**选择最多产的10位导演（电影数量最多的），绘制他们排行前3的三部电影的票房情况，并简要进行分析。

```python
Top_10_director = movie_data['director'].value_counts().index[:10]  # 电影数量最多的前10位导演

datas = movie_data['director'].isin(Top_10_director)
#movie_data[datas].grou['revenue_adj']#.sort_values(by=['revenue_adj'],ascending=False)[:3]
##movie_data.groupby('director').revenue_adj.agg(sum)
```

```python
'''已修改'''

Top_10_director = movie_data['director'].value_counts().index[:10]  # 电影数量最多的前10位导演

data = []
for d in Top_10_director:
    data.append(movie_data[['director','revenue_adj','original_title','profit']]
                .loc[ movie_data['director'] == d ]                    #提取前10位导演的所有电影
                .sort_values(by=['revenue_adj'],ascending=False)[:3])  #按票房排序并提取排行前3的电影                                    
'''
这个不知道要怎么修改，试了很多方法都会出错。。。
'''     
datas = pd.DataFrame()
datas = datas.append(data) 

Top_10_director = datas.groupby('director').revenue_adj.agg('sum').sort_values(ascending = False).index  # 电影数量最多的前10位导演,并排序

'''簇状柱形图'''
plt.figure(figsize=[20,30])
plt.subplot(3,1,1)
sb.barplot(data = datas,y = 'director',x = 'revenue_adj',hue = 'original_title',order=Top_10_director,palette='muted',dodge=False)
plt.xlabel('Revenue  /1 billion')  

'''每位导演的总票房条形图'''
plt.subplot(3,1,2)
for i in range(10):
    plt.bar(datas[ datas['director']==Top_10_director[i] ]['director'].values,
           datas[ datas['director']==Top_10_director[i] ]['revenue_adj'].values.sum(),color='green')
plt.ylabel('Revenue  /1 billion')

'''
每个导演电影的平均利润
'''
plt.subplot(3,1,3)
for i in range(10):
    plt.bar(datas[ datas['director']==Top_10_director[i] ]['director'].values,
           datas[ datas['director']==Top_10_director[i] ]['profit'].values.mean(),color='green')
plt.ylabel('Average profit')

'''
分析：
通过2个图，可以看出
1.Steven Spielberg的3部电影的总票房最高
2.排行前10的导演，电影平均利润和票房成正比
''';
```

**[选做]任务3.4：**分析1968年~2015年六月电影的数量的变化。

```python
'''已修改'''

june = pd.to_datetime(movie_data['release_date']).dt.month == 6  # 提取所有6月份的时间
year = movie_data['release_year'].between(1968,2015)   # 筛选1968-2015年的电影数量
datas = movie_data[year&june]['release_year'].value_counts().sort_index()

plt.figure(figsize=[20,5])
plt.plot(datas.index, datas.values)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Counts', fontsize=12)

'''
由图表可知，六月份发行的电影数量，从20世纪开始爆发性增长，之后一直处于增长状态
'''
```

**[选做]任务3.5：**分析1968年~2015年六月电影 `Comedy` 和 `Drama` 两类电影的数量的变化。

```python
df_genres = movie_data.drop('genres', axis=1).join(movie_data['genres'].str.split('|', expand=True).stack().reset_index(level=1, drop=True).rename('genres'))

# 筛选条件1 年份
sel_year = df_genres['release_year'].between(1968, 2015)

# 筛选条件2 六月
sel_June = pd.to_datetime(df_genres['release_date']).dt.month == 6

# 筛选条件3 类型
sel_genre = df_genres['genres'].isin(['Drama', 'Comedy'])
drama_genre = df_genres['genres'].isin(['Drama'])
comedy_genre = df_genres['genres'].isin(['Comedy'])
# 筛选数据并作图(参考逻辑读取部分)
plt.figure(figsize=[18, 5])
sb.countplot(data=df_genres[sel_year&sel_June&sel_genre], x='release_year', hue='genres')
plt.xticks(rotation=90)


'''
由图标可知
1.Comedy比Drama类型的电影发展的早
2.Comedy类型的数量在2009年达到巅峰，之后Drama开始批量出现，趋于稳定
''';
```
